##### Combining datasets education dataset and its meta data

```python
#Let us combine region and indicators to explore the markert available in each region.
regions=pd.read_csv('EdStatsCountry.csv')
region_incom=regions[['Country Code','Region','Income Group']]
education_region_combined=pd.merge(education,region_incom,on='Country Code',how='inner')
education_combined=pd.merge(education,region_incom,on='Country Code',how='inner')
```

#### students enrolled in secondary school

```python
# students enrolled in secondary school
StudentEnrolled_secondary=education_combined[education_combined['Indicator Code']=='SE.SEC.ENRL'].drop(['Unnamed: 69'],axis=1)
# since the the above dataframe are extracted from another dataframe we need to need to reset their index
StudentEnrolled_secondary.reset_index(drop=True, inplace=True)

# for making easy to visualize let's make the column of years and their values in another column
StudentEnrolled_secondary_re=pd.melt(StudentEnrolled_secondary,id_vars=(['Country Name','Country Code','Indicator Name','Indicator Code','Region','Income Group']),var_name='Year',value_name='secondary_students')
StudentEnrolled_secondary_indexed=StudentEnrolled_secondary_re.set_index('Year')
StudentEnrolled_secondary_indexed.index=pd.to_datetime(StudentEnrolled_secondary_indexed.index)
StudentEnrolled_secondary_selected=StudentEnrolled_secondary_indexed['2000':'2017']

#Let'us drop non-values in secondary_students column and rename Country Name to CountryName to easy our analysis
StudentEnrolled_secondary_selected1=StudentEnrolled_secondary_selected.dropna(subset=['secondary_students'])
StudentEnrolled_secondary_selected1.rename(columns={'Country Name':'CountryName'},inplace=True)
```

```python
# Normalizing our values in secondary_students column to be in range from 0 to 1
#using min-max normalization
StudentEnrolled_secondary_selected1['secondary_students']=((StudentEnrolled_secondary_selected1['secondary_students']-StudentEnrolled_secondary_selected1['secondary_students'].min())/(StudentEnrolled_secondary_selected1['secondary_students'].max()-StudentEnrolled_secondary_selected1['secondary_students'].min()))

StudentEnrolled_secondary_selected2=StudentEnrolled_secondary_selected1[StudentEnrolled_secondary_selected1['secondary_students']>=0.3]
plt.figure(figsize = (15,8))
sns.barplot(x='CountryName',y='secondary_students',data=StudentEnrolled_secondary_selected2)
plt.title('Ranking top five country with number of secondary students scale of 1 ')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize = (15,8))
plt.style.use('fivethirtyeight')
sns.lineplot(x='Year',y='secondary_students',hue='CountryName',data=StudentEnrolled_secondary_selected2)
plt.title('Enrolled student in secondary school from 2000 to 2017')
# plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.style.use('fivethirtyeight')
sns.lineplot(x='Year',y='secondary_students',hue='Region',data=StudentEnrolled_secondary_selected2)
plt.title('Enrolled student in secondary school from 2000 to 2017 by region')
# plt.legend(loc='upper left')
plt.show()
```

The data on ernorlled student is biased it can be hard to draw a conclusion from their ,This is because for example on the barchart above we were expecting top 5 countries with higher number of registered secondary students but we got income groups.This dipict that the data correction was some how biased on that case.

#### Internet per  100 people

```python
#Internet per  100 people
internet_per100=education_combined[education_combined['Indicator Code']=='IT.NET.USER.P2'].drop(['Unnamed: 69'],axis=1)

internet_per100.reset_index(drop=True, inplace=True)

# for making easy to visualize let's make the column of years and their values in another column internet per 100
internet_per100_new=pd.melt(internet_per100,id_vars=(['Country Name','Country Code','Indicator Name','Indicator Code','Region','Income Group']),var_name='Year',value_name='internet_per100')
internet_per100_indexed=internet_per100_new.set_index('Year')
internet_per100_indexed.index=pd.to_datetime(internet_per100_indexed.index)
internet_per100_selected=internet_per100_indexed['2000':'2017']

internet_per100_selected1=internet_per100_selected.dropna(subset=['internet_per100'])
internet_per100_selected1.rename(columns={'Country Name':'CountryName'},inplace=True)

internet_per100_selected1['internet_per100']=((internet_per100_selected1['internet_per100']-internet_per100_selected1['internet_per100'].min())/(internet_per100_selected1['internet_per100'].max()-internet_per100_selected1['internet_per100'].min()))

```

```python
internet_per100_selected2=internet_per100_selected1[internet_per100_selected1['internet_per100']>=0.99]
plt.figure(figsize = (15,8))
sns.barplot(x='CountryName',y='internet_per100',data=internet_per100_selected2)
plt.title('Ranking top five country in internet access per 100 people on scale of 1 ')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize = (15,8))
plt.style.use('fivethirtyeight')
sns.lineplot(x='Year',y='internet_per100',hue='CountryName',data=internet_per100_selected2)
plt.title('Internet per 100 people')
# plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.style.use('fivethirtyeight')
sns.lineplot(x='Year',y='internet_per100',hue='Region',data=internet_per100_selected2)
plt.title('Internet per 100 people')
# plt.legend(loc='upper left')
plt.show()
```

The bar graph above shows the top ten countries with many people with access to internet in 100 people.Bermuda was ranked number one with other countries like Iceland,Andorra,Bahrain,Liechtenstein,and Luxembourg arround it. The line graph shows Iceland increasing from 2014 up 2017.for regions North America and middle east& North Africa showed up as among top 5 regions with higer number of access to internet but in all of these two regions the number was decreasing.To sum up these data is also biased,so its difficult to draw a conclusion from these. 

#### personal computer per 100 population
